```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
import cv2
import tensorflow
from keras.models import Sequential
from keras.layers import Conv2D, Flatten, MaxPool2D, Dense
```

```
x=[]
y=[]
path="/content/train"
for p in os.listdir(path):
  if p.startswith("cat"):
    category=1
  else:
    category=0
  img_arr=cv2.imread(os.path.join(path+"/",p))
  newimg_arr=cv2.resize(img_arr,(128,128))
  x.append(newimg_arr)
  y.append(category)
```

```
path1="/content/test1"
xtest=[]
for p in os.listdir(path1):
  img_arr=cv2.imread(os.path.join(path1+"/",p))
  newimg_arr=cv2.resize(img_arr,(128,128))
  xtest.append(newimg_arr)
```

```
newx=np.array(x)
newx=newx/255
```

```
newtest=np.array(xtest)
newtest=newtest/255
```

```
newy=np.array(y)
```

```
model=Sequential()
model.add(Conv2D(64,(3,3),activation="relu",input_shape=newx.shape[1:]))
model.add(MaxPool2D(2,2))
model.add(Conv2D(64,(3,3),activation="relu"))
model.add(MaxPool2D(2,2))
model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
model.summary()
```

```
model.fit(newx,newy,epochs=5,batch_size=32,validation_split=0.2)
```

```
prediction=model.predict(newtest)
```

```
predicted_val = [int(round(p[0])) for p in prediction]
```
